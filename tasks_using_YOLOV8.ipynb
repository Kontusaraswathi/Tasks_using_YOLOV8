# tracking
track() → Adds object tracking, uses object IDs across frames, and needs persist=True.


```python
model=YOLO("yolov8l.pt")
'''You load a pre-trained YOLOv8 nano model (yolov8n.pt). The n stands for nano, which is the smallest version – faster but less accurate.

Alternatives:

yolov8s.pt → Small

yolov8m.pt → Medium

yolov8l.pt → Large

yolov8x.pt → eXtra large (most accurate, slowest)

'''
```

```python
cap=cv2.VideoCapture("vecteezy_black-and-white-horses-running-freely_2020707.mov")
ret=True
while ret:
    ret,frame=cap.read()
    results=model.track(frame,persist=True)
    frame_=results[0].plot()
    cv2.imshow("frame",frame_)
    if cv2.waitKey(1)&0XFF==ord('q'):
        break
```

```python
CAP=cv2.VideoCapture(0)
while True:
    ret,frame=CAP.read()
    results=model.track(frame,persist=True)
    frame_=results[0].plot()
    cv2.imshow("frame",frame_)
    if cv2.waitKey(1)&0XFF==ord('q'):
        break
CAP.release()
cv2.destroyAllWindows()
```

# detection
predict() → For detection only, each frame is treated independently.

```python
CAP=cv2.VideoCapture(0)
while True:
    ret,frame=CAP.read()
    results=model.predict(frame)
    frame_=results[0].plot()
    cv2.imshow("frame",frame_)
    if cv2.waitKey(1)&0XFF==ord('q'):
        break
CAP.release()
cv2.destroyAllWindows()
```

tip: Want faster detection?
Use yolov8n.pt for speed (less accurate), or try yolov8s.pt for a balance.

# Segmentation
It draws bounding boxes + masks (shaded areas) around each detected object.

```python
from ultralytics import YOLO
import cv2

model=YOLO("yolov8n-seg.pt")

cap=cv2.VideoCapture(0)
while True:
    ret,frame=cap.read()
    # segmentation
    results=model.predict(frame)
    frame_=results[0].plot()

    cv2.imshow("segmentation",frame_)
    if cv2.waitKey(1)&0XFF==ord('q'):
        break
cap.release()
cv2.destroyAllWindows()
```
